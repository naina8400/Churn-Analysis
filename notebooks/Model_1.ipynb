```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
import warnings
warnings.filterwarnings(action='ignore')
```

```python
base_dfz = pd.read_csv("tel_churn1.csv")
base_dfz.shape
```

```python
base_dfz
```

```python
base_dfz.columns
```

```python
base_dfz.TotalCharges = pd.to_numeric(base_dfz.TotalCharges, errors='coerce')
base_dfz.isnull().sum()
```

```python
base_dfz.dropna(how = 'any', inplace = True)
```

```python
base_dfz.shape
```

```python
base_dfz=base_dfz.drop('Unnamed: 0',axis=1)
```

```python
x=base_dfz.drop('Churn',axis=1)
x.shape
```

```python
y=base_dfz['Churn']
y.shape
```

```python
#train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Decision Tree Classifier

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

```python
model_dt_smote=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt_smote.fit(xr_train,yr_train)
yr_predict = model_dt_smote.predict(xr_test)
model_score_r = model_dt_smote.score(xr_test, yr_test)
print(model_score_r)
print(metrics.classification_report(yr_test, yr_predict))
```

```python
print(metrics.confusion_matrix(yr_test, yr_predict))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf.fit(x_train,y_train)
```

```python
y_pred=model_rf.predict(x_test)
```

```python
model_rf.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```
